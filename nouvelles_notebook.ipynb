```python
import pandas
import csv
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
nouvelles = pandas.read_csv("nouvelles.csv", sep=";", index_col=False)
nouvelles['index'] = range(1, len(nouvelles) + 1)
nouvelles = nouvelles.set_index("index")
nouvelles["annee"] = nouvelles["date"].map(lambda x: x[-4:])
nouvelles["check"] = "❌"
nouvelles.loc[nouvelles["lu"] == True, "check"] = "👌"
pandas.set_option('display.max_rows', len(nouvelles))
nouvelles
```

```python
nouvelles["lu"].sum(), len(nouvelles["lu"]), nouvelles["lu"].mean() * 100
```

```python
plt.plot(nouvelles["annee"].value_counts().sort_index())
plt.plot(nouvelles[nouvelles["lu"] == True]["annee"].value_counts().sort_index())
plt.xticks(rotation=-45)
```

```python
nouvelles.groupby("annee")["lu"].mean()
```

```python
nouvelles.groupby("annee")["lu"].count()
```

```python
nouvelles.to_csv("nouvelles_modified.csv", quoting=csv.QUOTE_NONNUMERIC)
# pandas.read_csv("nouvelles_modified.csv", index_col=0)
```
